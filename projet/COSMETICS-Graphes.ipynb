```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Graphe 1 : LINEPLOT 
vérifie la relation linéaire entre 2 variables

```python
#détermine valeurs abscisses
x = np.arange(0,50)

#création plot
plt.plot(x, x/3 , color="C3")

#on met en forme
plt.ylabel('Axe des y')
plt.xlabel('Axe des x')
plt.title('Tracer une droite')

#on affiche
plt.show()
```

## Upload CSV

```python
data = pd.read_csv('company_sales_data.csv',  sep=',')
data
```

## Graphe 2 : PROFIT TOTAL REALISE CHAQUE MOIS

```python
#déterminer taille matrice
fig, axe = plt.subplots(figsize=(10,8))


#détermine valeurs abscisses
x = np.arange(1,13)

#détermine valeurs ordonnées
y = data.iloc[0:12]['total_profit']

#création plot
plt.plot(x, y , color="C10")

#paramètre les axes
plt.axis([0,13,150000,450000])

#on met en forme
plt.ylabel('Profit')
plt.xlabel('Months')
plt.title('Total Profit per Month')


#ajout de value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])

addlabels(x,y)



#on affiche
plt.show()
```

## Graphe 3 : PROFIT  PAR MOIS POUR CHAQUE PRODUIT

```python
#déterminer taille matrice
fig, axe = plt.subplots(figsize=(14,8))


#détermine valeurs abscisses
x = np.arange(1,13)

#détermine valeurs ordonnées
y1 = data.iloc[0:12]['facecream']
y2 = data.iloc[0:12]['facewash']
y3 = data.iloc[0:12]['toothpaste']
y4 = data.iloc[0:12]['bathingsoap']
y5 = data.iloc[0:12]['shampoo']
y6 = data.iloc[0:12]['moisturizer']

#création plot
plt.plot(x, y1 , color="pink", label="facecream")
plt.plot(x, y2 , color="blue", label="facewash")
plt.plot(x, y3 , color="green", label="toothpaste")
plt.plot(x, y4 , color="yellow", label="bathingsoap")
plt.plot(x, y5 , color="purple", label="shampoo")
plt.plot(x, y6 , color="grey", label="moisturizer")

#paramètre les axes
plt.axis([0,13,0,15000])

#on met en forme
plt.ylabel('Profit')
plt.xlabel('Months')
plt.title('Total Profit for each Category of Product per Month')

#légende
plt.legend(loc="upper left", title="Produits", frameon=False)


#on affiche
plt.show()
```

## Graphe 4 : PRODUIT LE PLUS VENDU CHAQUE MOIS

```python
#je sélectionne les colonnes du tableau qui m'intéressent et renomme l'index
df = data.iloc[0:12][['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]
df = df.rename(index={0:'Jan', 1:'Feb', 2:'Mar', 3:'Apr', 4:'May', 5:'Jun', 6:'Jul', 7:'Aug',8:'Sep',9:'Oct',10:'Nov',11:'Dec'})

#je crée un graphe avec barres empilées (stacked)
df.plot(kind="bar", stacked =True, figsize=(15,8))

#mise en forme
plt.ylabel('Profit')
plt.xlabel('Months')
plt.title('Total Profit for each product each month')
```

## Graphe 5 : CONTRIBUTION DE CHAQUE PRODUIT DANS LE PROFIT TOTAL

```python
#définir liste produits
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']

#définir liste des sommes
data_pie = [data['facecream'].sum(),data['facewash'].sum(),data['toothpaste'].sum(),data['bathingsoap'].sum(),data['shampoo'].sum(),data['moisturizer'].sum(),]


#creating plot
plt.axis("equal")
plt.pie(data_pie, labels = products, autopct='%1.1f%%')


#mettre en forme
plt.legend(loc='lower right')
plt.title(' percentage profit for each product')
plt.show()
```
